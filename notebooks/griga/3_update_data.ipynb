```python
import pandas as pd
import requests
from tqdm import tqdm
import numpy as np
```

```python
data = pd.read_csv('data/india_clf_6.csv')[['Id', 'Body', 'Target']].set_index('Id')
data = data.rename(columns={
    'Id': 'id',
    'Body': 'description',
    'Target': 'class'
})
data['subject'] = 'no subject'
data
```

```python
# Находим дубликаты по описанию и оставляем только записи с наиболее частым классом
def remove_duplicates_keep_most_frequent_class(data):
    # Группируем по описанию и находим наиболее частый класс для каждого
    most_frequent_class = data.groupby('description')['class'].agg(
        lambda x: x.value_counts().index[0]
    ).reset_index()
    
    # Объединяем с исходными данными, чтобы получить полные записи
    result = data.merge(
        most_frequent_class, 
        on=['description', 'class'], 
        how='inner'
    )
    
    return result

# Применяем функцию
data_cleaned = remove_duplicates_keep_most_frequent_class(data)

# Проверяем результат
print(f"Исходное количество записей: {len(data)}")
print(f"Количество записей после удаления дубликатов: {len(data_cleaned)}")
print(f"Удалено дубликатов: {len(data) - len(data_cleaned)}")

# Проверяем, что дубликаты действительно удалены
duplicates_after = data_cleaned['description'].value_counts()
print(f"\nКоличество описаний, которые встречаются больше одного раза: {(duplicates_after > 1).sum()}")
data_cleaned.loc[data_cleaned['class'] == 'Others | General query', 'class'] = 'Others'
# Сохраняем очищенные данные
data_cleaned.to_csv('data/india_clf_6_cleaned.csv')
```
